```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure the DataFrame is correctly referenced
sns.pairplot(train[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']], hue='Survived', palette='coolwarm')

# Display the plot
plt.suptitle('Pairplot of Numerical Features by Survival', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size for better readability
sns.boxplot(x="Survived", y="Fare", data=train)
plt.xlabel("Survival (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.title("Boxplot of Fare by Survival Status")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('../Dataset/train.csv')

# Compute correlation matrix
correlation_matrix = df[["Survived", "Fare", "Age", "SibSp", "Parch"]].corr()

# Create and display the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap of Numerical Feature Correlations")
plt.show()
```

```python
# Select numerical features
numeric_features = ["Age", "Fare", "SibSp", "Parch"]

# Create histograms
plt.figure(figsize=(12, 8))  # Adjust size for better readability
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(2, 2, i)  # Arrange plots in a 2x2 grid
    plt.hist(df[feature].dropna(), bins=30, color="skyblue", edgecolor="black")  # Remove NaN values
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Count")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# Create scatterplot (Example: Fare vs. Age)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df, palette="coolwarm", alpha=0.7)

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatterplot of Age vs. Fare (Colored by Survival)")
plt.legend(title="Survival", labels=["Not Survived", "Survived"])
plt.show()
```
